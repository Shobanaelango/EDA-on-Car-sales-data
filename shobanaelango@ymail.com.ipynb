# 	6.1.Top 10 Most Purchased Car brands

```python
car.car.value_counts().nlargest(10).sort_values().plot(kind = 'bar',figsize=(10, 6), color='purple', grid=False)
plt.xlabel('Cars')
plt.ylabel('Number of Cars Purchased')
```

We can see from the plot that the most preferred car brand of all time is Volkswagen and the others are the top 10 most preferred models

<a id = Section6.2></a>
# 	6.2.Most preferred model from the top 10 car brands

```python
model_car_freq =car.groupby(['car','model']).size().sort_values(ascending=False).head(10).reset_index(name='Freq')
print (model_car_freq)
model_car_freq['model_car'] = model_car_freq[['car','model']].apply(lambda x: '/'.join(x), axis=1)
model_car_freq

plt.figure(figsize=(15, 10))
sns.barplot(x='model_car', y='Freq', data=model_car_freq)
plt.xticks(rotation=90)
plt.tight_layout()
#bowler_economy.head()
#for i,v in enumerate(model_car_freq.values): #write actual values on bar
#    plt.text(x=i, y=v[1]+0.15, s=round(v[1],2))

plt.xlabel('car')
plt.ylabel('Number of Car model combinations Purchased')
```

Mercedes-Benz and E-Class   is the most purchased Car brand model by users

<a id = Section6.3></a>
# 6.3.Top 10 highest price of cars bought by users.

```python


price_of_cars= car.groupby('car')['price'].sum().reset_index().sort_values(by='price', ascending = False)
df = price_of_cars.sort_values('price',ascending = False).groupby('price').head(5)
df = df.nlargest(10, 'price')
print(df)
plt.figure(figsize=(15, 10))
sns.barplot(x='car', y='price', data=df)
plt.xticks(rotation=90)
plt.tight_layout()
#bowler_economy.head()
for i,v in enumerate(df.values): #write actual values on bar
    plt.text(x=i, y=v[1]+0.15, s=round(v[1],2))

plt.xlabel('Car')
plt.ylabel('Price of Cars Purchased')
```

Mercedes-Benz is the costliest car preferred most by users

# 6.5.Cars registration  pattern through years

```python

yes_registration = car[car["registration"]=="yes"]
no_registration = car[car["registration"]=="no"]
# print(yes_registration)
# print(no_registration)
yes_registration_join = yes_registration.groupby(['car','year']).size().sort_values(ascending=False).reset_index(name='Freq')
no_registration_join = no_registration.groupby(['car','year']).size().sort_values(ascending=False).reset_index(name='Freq')

print(yes_registration_join)
print(no_registration_join)



# yes_registration.groupby(['car'])['year'].mean().plot(kind='line', figsize=(15, 7), color='green')
# no_registration.groupby(['car'])['year'].mean().plot(kind='line', figsize=(15, 7), color='red')

yes_registration_join.groupby(['year'])['Freq'].mean().plot(kind='line', figsize=(15, 7), color='green')
no_registration_join.groupby(['year'])['Freq'].mean().plot(kind='line', figsize=(15, 7), color='red')


plt.ylabel('Number of cars')
plt.title('Registered and UnRegistered Cars through years')
plt.legend(['Registered', 'UnRegistered'])
```

# 6.7.Most preferred body type in the past 10 years

```python

from datetime import date

today = date.today()
print("Today's year:", today.year)

no_reg = car[car["year"]> (today.year-10)]
no_reg
no_reg.body.unique()
body_type = no_reg.body.value_counts()
body_type


body_type =no_reg.groupby(['body']).size().sort_values(ascending=False).head(100).reset_index(name='Count')
print (body_type)
#body_type['model_car'] = model_car_freq[['car','model']].apply(lambda x: '/'.join(x), axis=1)
#model_car_freq

#body_type.body.value_counts().nlargest().sort_values().plot(kind = 'bar',figsize=(10, 7), color='blue', grid=True)


plt.figure(figsize=(10, 7))
sns.barplot(x='body', y='Count', data=body_type)
plt.xticks(rotation=90)
plt.tight_layout()

plt.xlabel('Car body type')
plt.ylabel('Car counts')

```

From the bar plots, we can see that crossover is the most preffered body type followed by Sedan

# 6.8.Top most prefered body type car brands of all time

# 6.11.Preference of people based on engine type

```python
car.engType.unique()

# Using pandas plot function to plot a donut plot of the Category column.
# Using kind='pie' to plot a donut plot, also setting explode values so that it takes a donut shape instead of a pie shape.

space = np.ones(4)/4
car['engType'].value_counts().plot(kind='pie', explode=space, fontsize=15, autopct='%3.1f%%', wedgeprops=dict(width=0.2), 
                                       shadow=True, startangle=160, figsize=(11,10), cmap='inferno', legend=True)

plt.ylabel('Engine Type ')
```
